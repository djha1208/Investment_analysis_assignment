# Sector Analysis 1

```python
# Extract the primary sector of each category list from the category_list column
master_frame['primary_sector'] = master_frame.category_list.apply(lambda x: x.split('|')[0])
master_frame
```

```python
# Importing mapping csv file here
mapping = pd.read_csv("mapping.csv", encoding = "ISO-8859-1")
mapping
```

```python
# different categories in mapping dataframe
print(set(mapping['category_list'])-set(master_frame['primary_sector']))
```

```python
# Replacing '0' in the category list column by 'na' 
# Replacing '2.na' in the company list column by '2.0'
mapping['category_list']=mapping['category_list'].str.replace('0','na')
mapping['category_list']=mapping['category_list'].str.replace('2.na','2.0')
mapping
```

```python
# Map each primary sector to one of the eight main sector
# Restructure the mapping table with only "category_list","main_sector" columns
# Renaming category_list as primary_sector
mapping['main_sector'] = mapping.iloc[:, 1:].idxmax(axis=1)
mapping = mapping[["category_list","main_sector"]]
mapping=mapping.rename(columns={'category_list':'primary_sector'})
mapping['primary_sector']=mapping['primary_sector'].str.lower()
mapping
```

```python
master_frame['primary_sector']=master_frame['primary_sector'].str.lower()
master_frame
```

```python
# inspecting master_frame for extra primary_sector which are not present in mapping
master_frame.loc[~master_frame['primary_sector'].isin(mapping['primary_sector']), :]
```

```python
# Merge master_frame and mapping using common column primary_sector (Inner join)
master_frame = pd.merge(master_frame,
                 mapping,
                 on='primary_sector',
                 how='inner')
master_frame
```

## Sector Analysis  2

```python
# Creating new data frame D1 of country 'USA' which contain funding type 'venture' having 5 to 15 million USD
# Creating new data frame D2 of country 'GBR' which contain funding type 'venture' having 5 to 15 million USD
# Creating new data frame D3 of country 'IND' which contain funding type 'venture' having 5 to 15 million USD
Country1 = "USA"
Country2 = "GBR"
Country3 = "IND"
FT = "venture"
D1 = pd.DataFrame(master_frame.loc[(master_frame.funding_round_type == FT) & (master_frame.raised_amount_usd >= 5000000.0) & (master_frame.raised_amount_usd <= 15000000.0) & (master_frame.country_code == Country1)])
D2 = pd.DataFrame(master_frame.loc[(master_frame.funding_round_type == FT) & (master_frame.raised_amount_usd >= 5000000.0) & (master_frame.raised_amount_usd <= 15000000.0) & (master_frame.country_code == Country2)])
D3 = pd.DataFrame(master_frame.loc[(master_frame.funding_round_type == FT) & (master_frame.raised_amount_usd >= 5000000.0) & (master_frame.raised_amount_usd <= 15000000.0) & (master_frame.country_code == Country3)])
print(D1.shape)
print(D2.shape)
print(D3.shape)
```

```python
print(D1.raised_amount_usd.sum())
print(D2.raised_amount_usd.sum())
print(D3.raised_amount_usd.sum())
```

```python
# Total number of investments in 'USA' per each main_sector
D1.pivot_table(values = 'raised_amount_usd', index = 'main_sector', aggfunc = 'count').sort_values(by = 'raised_amount_usd', ascending = False)
```

```python
# Total number of investments in 'GBR' per each main_sector
D2.pivot_table(values = 'raised_amount_usd', index = 'main_sector', aggfunc = 'count').sort_values(by = 'raised_amount_usd', ascending = False)
```

```python
# Total number of investments in 'IND' per each main_sector
D3.pivot_table(values = 'raised_amount_usd', index = 'main_sector', aggfunc = 'count').sort_values(by = 'raised_amount_usd', ascending = False)
```

```python
# Total amount of investment in each company per main_sector in decreasing order for 'USA'
D1.pivot_table(values = 'raised_amount_usd', index = ['main_sector', "name"], aggfunc = 'sum').sort_values(by = 'raised_amount_usd', ascending = False)
```

```python
# Total amount of investment in each company per main_sector in decreasing order for 'GBR'
D2.pivot_table(values = 'raised_amount_usd', index = ['main_sector', "name"], aggfunc = 'sum').sort_values(by = 'raised_amount_usd', ascending = False)
```

```python
# Total amount of investment in each company per main_sector in decreasing order for 'IND'
D3.pivot_table(values = 'raised_amount_usd', index = ['main_sector', "name"], aggfunc = 'sum').sort_values(by = 'raised_amount_usd', ascending = False)
```
